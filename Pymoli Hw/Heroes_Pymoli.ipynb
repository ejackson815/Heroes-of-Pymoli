```python
#Purchasing Analysis (Gender)
```

```python
female_group = df.loc[df["Gender"]=="Female"]
male_group = df.loc[df["Gender"]=="Male"]
non = df.loc[df["Gender"]=="Other / Non-Disclosed"]

#purchase count
fpc = len(female_group['Item ID'])
mpc = len(male_group['Item ID'])
npc = len(non['Item ID'])

#Average Purchase Price
fapp = sum(female_group['Price']) / len(female_group['Price'])
mapp = sum(male_group['Price']) / len(male_group['Price'])
napp = sum(non['Price']) / len(non['Price'])

#Total Purchase Price
ftpp =sum(female_group['Price'])
mtpp =sum(male_group['Price'])
ntpp =sum(non['Price'])
#Normalized Totals
fnt = ftpp / fpc
mnt = mtpp / mpc
nnt = ntpp / npc
```

```python
d = {'Gender' : ['Male', 'Female', "Other / Non-Disclosed"], 'Purchase Count': [mpc, fpc, npc], 'Average Purchase Price':[mapp, fapp, napp], 'Total Purchase Price':[mtpp, ftpp, ntpp],'Normalized Totals':[mnt, fnt, nnt] }
purch = pd.DataFrame(data=d)
purch = purch.round(2)
purch = purch[['Gender', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Price', 'Normalized Totals']]
purch
```
